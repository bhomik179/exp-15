```python
import pandas as pd
import numpy as np
data = {
    'Product': ['Laptop', 'Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
    'Price': [55000, 20000, 30000, 2000, 8000, 45000],
    'Units_Sold': [25, 60, 35, 80, 50, 15],
    'Discount':[10,15,12,20,18,8]
}
df = pd.DataFrame(data)
print(df)
```

```python
#1.Min-Max Normalization
# Min-Max Normalization for price
df['Price_MinMax'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())
print(df[['Product', 'Price', 'Price_MinMax']])
```

```python
#1.Min-Max Normalization
# Min-Max Normalization for Discount
df['Discount_MinMax'] = (df['Discount'] - df['Discount'].min()) / (df['Discount'].max() - df['Discount'].min())
print(df[['Product', 'Discount', 'Discount_MinMax']])
```

```python
# #1.Min-Max Normalization
# Min-Max Normalization for Unit_Sold
df['Units_Sold_MinMax'] = (df['Units_Sold'] - df['Units_Sold'].min()) / (df['Units_Sold'].max() - df['Units_Sold'].min())
print(df[['Product', 'Units_Sold', 'Units_Sold_MinMax']])
```

```python
#Normalize Multiple Columns
colsd=['Price','Units_Sold','Discount']
df[colsd]=(df[colsd]-df[colsd].min())/(df[colsd].max()-df[colsd].min())
print(df)
```

```python
#2.Z-score Normalization
df['Units_Sold_Zscore'] = (df['Units_Sold'] - df['Units_Sold'].mean()) / df['Units_Sold'].std()
print(df[['Product', 'Units_Sold', 'Units_Sold_Zscore']])
```

```python
#3. Decimal Scaling
df['Price_Decimal'] = df['Price'].apply(lambda x: round(x, 2))
print(df[['Product', 'Price', 'Price_Decimal']])
```

```python
#Decimal Scaling (Review/Price)
# Formula: x / 10^j (where j is the number of digits of the max absolute value)
def decimal_scaling(series):
    j = np.ceil(np.log10(series.abs().max()))
    return series / (10**j)

df_amazon['Price_scaled'] = decimal_scaling(df_amazon['Price']) # Replace 'price' with your actual column name
print(df_amazon)
```
